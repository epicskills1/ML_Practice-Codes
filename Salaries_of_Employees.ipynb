**QUESTION**
This hiring.csv file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates

```python
!pip install word2number
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df=pd.read_csv("hiring.csv")
df
```

```python
df.experience=df.experience.fillna("zero")
df
```

```python
# Convert all values in the 'experience' column to strings before applying word_to_num
df.experience = df.experience.astype(str).apply(w2n.word_to_num)
df
```

```python
import math
median_test_score = math.floor(df['test_score(out of 10)'].mean())
median_test_score
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[1,9,7]])
```
